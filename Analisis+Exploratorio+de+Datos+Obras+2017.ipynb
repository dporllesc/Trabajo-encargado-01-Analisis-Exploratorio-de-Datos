Trabajo encargado 01: Analisis Exploratorio de Datos

Carga de datos

```r
obra <- read.csv("obras2017_0.csv", header = TRUE, fileEncoding="latin1")
```

Revisar los sesis primeros datos por defecto

```r
head(obra)
```

Verificar la estructura de los datos

```r
str(obra)
```

Revisar un resumen de los datos

```r
summary(obra)
```

Revisar un resumen de los datos en columnas

```r
library(mlr)
summarizeColumns(obra)
```

Revisar graficamente el porcentaje de nulos

```r
library(VIM)
aggr(obra,numbers=TRUE, plot = T)
```

```r
library(gmodels)
library(gdata)
CrossTable(obra$REGION, format="SAS")
```

```r
CrossTable(obra$REGION, format="SPSS")
```

Graficos de resumen de variables: Grafico de Barras

```r
frec <- table(obra$REGION)
barplot(frec, main="Distribución de Regiones de Obras", xlab="Region", ylab="Cantidad de Obras")
```

Visualizar gráficos de resumen de las variables: pie

```r
pie(frec, main="Distribución de Region de obras", xlab="REGION")
```
