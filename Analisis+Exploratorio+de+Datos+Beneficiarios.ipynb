Trabajo encargado 01: Analisis Exploratorio de Datos

Carga de Datos

```r
consolidado <- read.csv("BeneficiariosConsolidado.csv", sep = ";", header = TRUE, fileEncoding="latin1")
```

Revisar los 6 primeros datos por defecto

```r
head(consolidado)
```

Verificar Estrucura de Datos

```r
str(consolidado)
```

Revisar un Resumen de datos

```r
summary(consolidado)
```

Revisar un resumen de datos en columnas

```r
library(mlr)
summarizeColumns(consolidado)
```

Revisar graficamente el porcentaje de nulos

```r
library(VIM)
aggr(consolidado,numbers=TRUE, plot = T)
```

 Visualizar la tabla y frecuencia con formato SAS

```r
library(gmodels)
library(gdata)
CrossTable(consolidado$DEPARTAMENTO, format="SAS")
```

Visualizar la tabla y frecuencia con formato SPSS


```r
frec <- table(consolidado$DEPARTAMENTO)
barplot(frec, main="Distribución del Departamento de los beneficirios", xlab="Departamento", ylab="Cantidad de Beneficiarios")
```

Visualizar gráficos de resumen de las variables: pie

```r
pie(frec, main="Distribución de departamento de los beneficiarios", xlab="Departamento")
```

Visualizar gráficos de resumen de las variables: pie 3D

```r
library(plotrix)
lbls1 <- paste(names(table(consolidado$DEPARTAMENTO)), "\n", prop.table(table(consolidado$DEPARTAMENTO))*100,"%", sep="") 
pie3D(table(consolidado$DEPARTAMENTO), labels = lbls1,explode=0.15, main="Distribución de Departamento de Beneficiarios")
```

Visualizar gráficos de resumen de las variables: grafico de puntos

```r
dotchart(table(consolidado$DEPARTAMENTO), cex=.7, main="Distribución de la Edad de los Clie", xlab="# de Clientes")
```
